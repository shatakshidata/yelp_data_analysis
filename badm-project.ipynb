```python
# Merge check-in data with restaurant data
merged_df = pd.merge(checkin_df, df_restaurant, on='business_id')

# Group by 'address', state and  count check-ins
checkin_counts = merged_df.groupby(['address', 'city', 'state']).size().reset_index(name='checkin_count')

# Sort by highest check-in counts
high_traffic_areas = checkin_counts.sort_values(by='checkin_count', ascending=False).head(20)

print("High Foot-Traffic Areas:")
print(high_traffic_areas)
```

```python
import matplotlib.pyplot as plt

# Plot the top 10 high-foot traffic areas
plt.figure(figsize=(20, 10))
top_areas = high_traffic_areas.head(50)
plt.barh(top_areas['address'], top_areas['checkin_count'], color='skyblue')
plt.xlabel('Check-In Count')
plt.ylabel('Address')
plt.title('Top 10 High Foot-Traffic Areas for Restaurants')
plt.gca().invert_yaxis()  # To display the highest count at the top
plt.show()
```

```python
# Merge check-in data with restaurant data to get location info
merged_df_new = pd.merge(checkin_df, df_restaurant, on='business_id')

# Aggregate check-ins by location (latitude and longitude)
checkin_counts = merged_df_new.groupby(['latitude', 'longitude']).size().reset_index(name='checkin_count')
```

```python
import folium
from folium.plugins import HeatMap

# Initialize the map centered around the average location
center_lat, center_long = checkin_counts['latitude'].mean(), checkin_counts['longitude'].mean()
m = folium.Map(location=[center_lat, center_long], zoom_start=12)

# Prepare data for the heatmap (latitude, longitude, check-in count as weight)
heat_data = [[row['latitude'], row['longitude'], row['checkin_count']] for index, row in checkin_counts.iterrows()]

# Add heatmap layer to the map
HeatMap(heat_data, radius=15, max_zoom=13).add_to(m)

# Display the map
m.save("checkin_heatmap.html")  # Save as HTML file if you want to view it in a browser
m
```

```python
philly_restaurants = business_df[
    (business_df['city'].str.contains('Philadelphia', case=False, na=False)) &
    (business_df['categories'].str.contains('Restaurant', case=False, na=False))
]
```

```python
# Merge Philadelphia restaurants with check-in data
philly_checkins = pd.merge(checkin_df, philly_restaurants[['business_id', 'categories']], on='business_id')
```

```python
# Split categories into lists and explode to get one row per cuisine
philly_checkins['categories'] = philly_checkins['categories'].str.split(', ')
philly_checkins_exploded = philly_checkins.explode('categories')

# Group by cuisine and count check-ins
cuisine_popularity = philly_checkins_exploded.groupby('categories').size().reset_index(name='checkin_count')

# Sort by check-in count in descending order to get the most popular cuisines
cuisine_popularity = cuisine_popularity.sort_values(by='checkin_count', ascending=False)

# Display top cuisines
print("Popular Cuisines in Philadelphia by Check-In Count:")
print(cuisine_popularity.head(10))
```

```python
import matplotlib.pyplot as plt

# Plot the top 10 cuisines by check-in count
top_cuisines = cuisine_popularity.head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_cuisines['categories'], top_cuisines['checkin_count'], color='skyblue')
plt.xlabel('Check-In Count')
plt.ylabel('Cuisine')
plt.title('Top 10 Popular Cuisines in Philadelphia by Check-In Count')
plt.gca().invert_yaxis()  # Highest check-in count at the top
plt.show()
```
